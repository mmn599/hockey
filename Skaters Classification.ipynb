### Get the data

```python
from sklearn.externals import joblib
from IPython.core import display as ICD

data = joblib.load('2015Skaters.p')
outputs = [x for x in data.columns.values if 'O_' in x]
ICD.display(data)
```

### Clean the data

Remove records with small Num (Num is the number of previous games the target player has played in this season). <br>
Remove the non numerical columns.

```python
num_threshold = 20
gnum_threshold = 10
data = data[data.Num >= num_threshold]
data = data[data.GNum >= gnum_threshold]
print(data['Num'].describe())
print(data['GNum'].describe())
```

```python
data = data.fillna(0)
```

```python
dropcols = ['GameNum', 'GameName', 'Player', 'DateTimestamp', 'Num']
cleandata = data.drop(dropcols, 1)
print(cleandata.columns.values)
```
